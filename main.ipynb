```python
import pandas as pd
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('산소산.csv')
data = data.dropna()
data
```

```python
data.columns
```

```python
x = data[['Oxygen', 'Electronegativity']]
x.shape
```

```python
y = data[['pH(0.1M)']]
y.shape
```

```python
X = tf.keras.layers.Input(shape=[2])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse',metrics=['accuracy'],optimizer='sgd')
```

```python
hist = model.fit(x,y,epochs=1000)
```

### 오차 그래프 (학습을 진행할수록)

```python
plt.plot(hist.history['loss'],label='loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

### 정확도 그래프 (학습을 진행할수록)

```python
plt.plot(hist.history['accuracy'],label='accuracy',color='orange')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```
